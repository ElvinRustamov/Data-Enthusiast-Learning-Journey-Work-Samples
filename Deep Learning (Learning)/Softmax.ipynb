```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
# "manually" in numpy

# the list numbers
z = [1, 2, 3]

# compute the softmax result
num = np.exp(z)
den = np.sum(np.exp(z))
sigma = num / den

print(sigma)
print(np.sum(sigma))
```

```python
# repeat with som random integers
z = np.random.randint(-5, high=15, size=15)
print(z)

# compute softmax result
num = np.exp(z)
den = np.sum(np.exp(z))
sigma = num / den

# compare
plt.plot(z, sigma, 'ko')
plt.xlabel("Original number (z)")
plt.ylabel("Softmax/sigma")
#plt.yscale('log')
plt.title('$\sum\sigma$ = %g' %np.sum(sigma))
plt.show()
```

```python
# using pytorch

# create an instance of the softmax activation class
softfun = nn.Softmax(dim=0)

# apply the data to that function
sigmaT = softfun(torch.Tensor(z))

print(sigmaT)
```

```python
plt.plot(sigma, sigmaT, 'ko')
plt.xlabel('"Manual" softmax')
plt.ylabel("Pytorch nn.Softmax")
plt.title(f"The two methods correlate at r={np.corrcoef(sigma, sigmaT)[0, 1]}")
```
